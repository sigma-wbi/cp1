**2.데이터 정제**


1.   foodtype 간결화 : foodtype 안의 데이터를 간결하고 직관적으로 바꿈 먼저 allData로 합치고 바꿔도 되는데 여기서는 따로 바꿈

````python
Nfusion = fusion.replace({'foodtype':"퓨전음식)"},'퓨전음식')
Nfusion = Nfusion.drop_duplicates()

Nbake = bake.replace({'foodtype':"아이스크림/도넛)"},'베이커리')
Nbake = Nbake.replace({'foodtype':"제과/베이커리)"},'베이커리')
Nbake = Nbake.drop_duplicates()

Nbunsic = bunsic.replace({'foodtype':"김밥/떡볶이/분식류)"},'분식')
Nbunsic = Nbunsic.drop_duplicates()

Nch = ch.replace({'foodtype':"중식/일식)"},'중식')
Nch = Nch.drop_duplicates()

Njp = jp.replace({'foodtype':"돈까스/우동/카레)"},'일식')
Njp = Njp.drop_duplicates()

Ncoffe = coffe.replace({'foodtype':"커피전문점/대형커피)"},'커피')
Ncoffe = Ncoffe.drop_duplicates()

Ndessert = dessert.replace({'foodtype':"디저트/브런치/샐러드)"},'디저트')
Ndessert = Ndessert.drop_duplicates()

Nmeat = meat.replace({'foodtype':"고기전문점)"},'고기')
Nmeat = Nmeat.replace({'foodtype':"족발/보쌈/곱창)"},'고기')
Nmeat = Nmeat.drop_duplicates()

Nrice = rice.replace({'foodtype':"죽/도시락/비빔밥)"},'한식')
Nrice = Nrice.drop_duplicates()

Nfast = fast.replace({'foodtype':"패스트푸드/피자)"},'패스트푸드')
Nfast = Nfast.drop_duplicates()

Nchicken = chicken.replace({'foodtype':"프 (치킨전문점)"},'치킨')
Nchicken = Nchicken.drop_duplicates()
````

위의 과정을 겪은 각각의 다른 종류의 음식테이블을 하나의 allData 데이터프레임으로합침

````python
allData = pd.concat([Nfusion,Nbake,Nbunsic,Nch,Njp,Ncoffe,Ndessert,Nmeat,Nrice,Nfast,Nchicken])
````

````python
allData = allData.reset_index()
allData = allData.drop('index',axis = 1)
````
